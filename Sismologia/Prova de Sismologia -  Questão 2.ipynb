Calcular o particionamento de energia devido a incidência de uma onda S numa superfície horizontal pode ser feito através dos potenciais para ondas planas. O potencial para uma onda S plana é dado por:

> $\Psi(x,t) = B exp[i(\omega t - k_{\beta} . x)]$

> $\Psi(x_{1},x_{2},x_{3},t) = B exp[i(\omega t - k_{1} . x_{1} - k_{2} . x_{2} - k_{3} . x_{3})]$

Para calcular os deslocamentos das partículas neste campo vetorial utiliza-se:

$u_{S} = \nabla \times \Psi = \frac{\partial \Psi _{3}}{\partial x_{2}} - \frac{\partial \Psi _{2}}{\partial x_{3}} \hat{x_{1}} + \frac{\partial \Psi _{1}}{\partial x_{3}} - \frac{\partial \Psi _{3}}{\partial x_{1}}\hat{x_{2}} +  \frac{\partial \Psi _{2}}{\partial x_{1}} - \frac{\partial \Psi _{1}}{\partial x_{2}}\hat{x_{3}}$

Como o plano de propagação escolhido foi o plano $x_{1}x_{3}$, vide gráfico abaixo, $k_{2}$ e $\frac{\partial \Psi}{\partial x_{2}}$ são iguais a zero. Logo pode-se calcular os deslocamentos através do potencial vetor para uma onda plana S:

> $u_{S_{1}} = - \frac{\partial \Psi _{2}}{\partial x_{3}} \hat{x_{1}}$

> $u_{S_{2}} = \frac{\partial \Psi _{1}}{\partial x_{3}} - \frac{\partial  \Psi _{3}}{\partial x_{1}} \hat{x_{2}}$

> $u_{S_{3}} = \frac{\partial \Psi _{2}}{\partial x_{1}} \hat{x_{3}}$


Sabendo que a componente $u_{S_{2}}$ do deslocamento da onda S representa a onda $SH$, temos:

> $u_{S_{2}} = u_{SH} = \frac{\partial \Psi _{1}}{\partial x_{3}} - \frac{\partial \Psi _{3}}{\partial x_{1}} \hat{x_{2}}$

sendo: 

> $\Psi(x,t) = B exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})]$

Por conseguinte:

> $\frac{\partial \Psi _{1}}{\partial x_{3}} = -ik_{3}B exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})]$

> $\frac{\partial \Psi _{3}}{\partial x_{1}} = -ik_{1}B exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})]$  


Como $u_{S_{2}} = u_{SH} = \frac{\partial \Psi _{1}}{\partial x_{3}} - \frac{\partial \Psi _{3}}{\partial x_{1}} \hat{x_{2}}$, então:

> $u_{SH} = -ik_{3}B exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})] - (-ik_{1}B exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})])$  

> $u_{SH} = (-ik_{3}B - (- ik_{1}B) )exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})]$

> $u_{SH} = (-ik_{3}B + ik_{1}B) exp[i(\omega t - k_{1} . x_{1} - k_{3} . x_{3})]$

sendo $\omega p = k_{1}$ e $\omega \eta _{\beta} = k_{3}$, manipulando matematicamente teremos:

> $u_{SH} = (-i\omega \eta _{\beta}B + i\omega pB) exp[i(\omega t - \omega p . x_{1} - \omega \eta _{\beta} . x_{3})]$

> $u_{SH} = i\omega B(p - \eta _{\beta}) exp[i\omega (t - p . x_{1} - \eta _{\beta} . x_{3})]$

>>> Para minizar os cálculos seguintes faz-se  $i\omega B(p - \eta _{\beta}) = C$.

Quando um onda $SH$ incide ($SH_{i}$) sob um discontinuidade horizontal à superfície são geradas apenas dois tipo de ondas: 

> (1) $SH$ refletida ($SH_{r}$);

> (2) $SH$ transmitida ($SH_{t}$).

Considerando $x_{3} < 0$ para a camada 1 e $x_{3} > 0$ para a camada 2. Então, para as ondas que estão no sentido positivo de $x_{3}$ as exponenciais recebem  $(t - p . x_{1} - \eta _{\beta} . x_{3})$, já as que estão no sentido negativo de $x_{3}$ os expoentes serão $(t - p . x_{1} + \eta _{\beta} . x_{3})$. Apenas questão de conversão. Portanto:

>> $SH_{r} = C_{r} exp[i\omega (t - p . x_{1} + \eta _{\beta 1} . x_{3})]$

>> $SH_{t} = C_{t} exp[i\omega (t - p . x_{1} - \eta _{\beta 2} . x_{3})]$

Segundo o desenho acima, os deslocamentos para a camada 1, $u_{1}$, são dados pela soma dos deslocamentos das ondas $SH$ incidente e refletida, $SH_{i}$ + $SH_{r}$, portanto:

$u_{1}(x_{1},x_{3},t) = SH_{i} + SH_{r}$

$u_{1}(x_{1},x_{3},t) = C_{i} exp[i\omega (t - p . x_{1} - \eta _{\beta1 } . x_{3})] + i\omega C_{r} exp[i\omega (t - p . x_{1} + \eta _{\beta 1} . x_{3})]$

Já para a camada 2:

$u_{2}(x_{1},x_{3},t) = SH_{t}$

$u_{2}(x_{1},x_{3},t) = C_{t} exp[i\omega (t - p . x_{1} - \eta _{\beta 2} . x_{3})]$

As condições de contorno para este tipo de caso onde uma onda $SH$ incide sobre um superfície em um meio sólido-sólido são que no limite ($x_{3} = 0$) haja:

> (1) continuidade do deslocamento ($u^{-}(x_{1},0,t) = u^{+}(x_{1},0,t)$);

> (2) continuidade da tração através de um plano horizontal ($\tau ^{-} _{23}(x_{1},0,t) = \tau ^{+} _{23}(x_{1},0,t)$);

A continuidade do deslocamento é calculada atráves da:

>> $u^{-}(x_{1},0,t) = u^{+}(x_{1},0,t)$

$u^{-}(x_{1},0,t) =C_{i} exp[i\omega (t - p . x_{1} - \eta _{\beta 1} . 0)] + C_{r} exp[i\omega (t - p . x_{1} + \eta _{\beta 1} . 0)]$

$u^{-}(x_{1},0,t) = C_{i} exp[i\omega (t - p . x_{1})] + C_{r} exp[i\omega (t - p . x_{1})]$

### $u^{-}(x_{1},0,t) = (C_{i} + C_{r}) exp[i\omega (t - p . x_{1})]$

$u^{+}(x_{1},0,t) = C_{t} exp[i\omega (t - p . x_{1} - \eta _{\beta 2} . 0)]$

### $u^{+}(x_{1},0,t) = C_{t}(p - \eta _{\beta 2}) exp[i\omega (t - p . x_{1})]$

____

Igualando $u^{-}(x_{1},0,t) = u^{+}(x_{1},0,t)$ temos:

$(B_{i} + B_{r}) exp[i\omega (t - p . x_{1})] = C_{t} exp[i\omega (t - p . x_{1})]$

Cancelando os termos comuns em ambos os lados teremos:

>> ### $C_{i} + C_{r} =  C_{t}$

O cálculo da tração é dada por:

>> $\tau _{ij} = \lambda \delta _{ij} \partial _{k} u_{k} + \mu (\partial _{j} u_{j} + \partial _{j} u_{i})$

Os deslocamentos nas ondas $SH$ em $x_{1}$ e $x_{3}$ são zero, logo os esforços nesses eixos são iguais a zero. E ara um vetor normal (0,0,1) à interface, vemos primeiro termo para calcular $\tau _{ij}$ vai para zero e que os únicos componentes diferentes de zero sejam:

> $\tau _{23} = \mu (\partial _{3} u_{SH}) = \mu (\frac{\partial u_{SH}}{\partial x_{3}})$


Então para calcular a continuidade da tração fazemos:

>> $\tau ^{-} _{23}(x_{1},0,t) = \tau ^{+} _{23}(x_{1},0,t)$

$\tau ^{-} _{23}(x_{1},0,t) = \mu _{1}( - i \omega \eta _{\beta 1} C_{i} exp[i\omega (t - p . x_{1} - \eta _{\beta 1} . 0)] + i\omega \eta _{\beta 1} C_{r} exp[i\omega (t - p . x_{1} + \eta _{\beta 1} . 0)])$

Manipulando matematicamente e colocando $exp[i\omega (t - p . x_{1})]$ em evidência: 

### $\tau ^{-} _{23}(x_{1},0,t) = i \omega \mu _{1} \eta _{\beta 1}(C_{r} - C_{i}) exp[i\omega (t - p . x_{1})] $

_____

$\tau ^{+} _{23}(x_{1},0,t) = \mu _{2} (- i \omega \eta _{\beta 2} C_{t} exp[i\omega (t - p . x_{1} - \eta _{\beta} . 0)])$

### $\tau ^{+} _{23}(x_{1},0,t) = - i \omega \mu _{2} \eta _{\beta 2} C_{t} exp[i\omega (t - p . x_{1})]$
____

Igualando $\tau ^{-} _{23}(x_{1},0,t) = \tau ^{+} _{23}(x_{1},0,t)$ temos:

$i \omega \mu _{1} \eta _{\beta 1}(C_{r} - C_{i}) exp[i\omega (t - p . x_{1})] = - i \omega \mu _{2} \eta _{\beta 2} C_{t} exp[i\omega (t - p . x_{1})]$

Cancelando os termos comuns em ambos os lados:

$\mu _{1} \eta _{\beta 1}(C_{r} - C_{i})  = - \mu _{2} \eta _{\beta 2} C_{t}$

Multiplicando ambos os lados por (-1) e reorganizando temos:


>> ### $(C_{i} - C_{r})  = \frac{C_{t} \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1}}$

________

Por fim chegamos a esse sistema de equações:

$\left\{\begin{matrix}
C_{i} + C_{r} =  C_{t} (1)\\ 
(C_{i} - C_{r})  = \frac{C_{t} \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1}} (2)
\end{matrix}\right.$


O coeficiente de reflexão ($R$) é dado por $\frac{C_{r}}{C_{i}}$ e coeficiente de transmissão ($T$) é dado por $\frac{C_{t}}{C_{i}}$, com isso, dividimos as equações (1) e (2) por $C_{i}$ em ambos os lados:

$\frac{C_{i}}{C_{i}} + \frac{C_{r}}{C_{i}} =  \frac{C_{t}}{C_{i}}$

$1 + R =  T$ (3)

$\frac{C_{i}}{C_{i}} - \frac{C_{r}}{C_{i}} = \frac{C_{t} \mu _{2} \eta _{\beta 2}}{C_{i}\mu _{1} \eta _{\beta 1}}$ 

$1 - R = \frac{T \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1}}$ 

$(1 - R)\mu _{1} \eta _{\beta 1} = T \mu _{2} \eta _{\beta 2}$ (4)

Substituindo a equação (3) em (4) teremos:

$(1 - R)\mu _{1} \eta _{\beta 1} = (1 + R) \mu _{2} \eta _{\beta 2}$ 

Desenvolvendo os cálculos:

$\mu _{1} \eta _{\beta 1} - R \mu _{1} \eta _{\beta 1} = R \mu _{2} \eta _{\beta 2} + \mu _{2} \eta _{\beta 2}$

$\mu _{1} \eta _{\beta 1} - \mu _{2} \eta _{\beta 2} = R \mu _{2} \eta _{\beta 2} + R \mu _{1} \eta _{\beta 1}$

$\mu _{1} \eta _{\beta 1} - \mu _{2} \eta _{\beta 2} = R (\mu _{2} \eta _{\beta 2} + \mu _{1} \eta _{\beta 1})$

$\frac{\mu _{1} \eta _{\beta 1} - \mu _{2} \eta _{\beta 2}}{\mu _{2} \eta _{\beta 2} + \mu _{1} \eta _{\beta 1}} = R$

Arrumando:

### $R = \frac{\mu _{1} \eta _{\beta 1} - \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1} + \mu _{2} \eta _{\beta 2}}$
____

Substituindo $R$ na equação (3) teremos:

$1 + \frac{\mu _{1} \eta _{\beta 1} - \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1} + \mu _{2} \eta _{\beta 2}}=  T$

$T = \frac{\mu _{1} \eta _{\beta 1} + \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1} + \mu _{2} \eta _{\beta 2}} + \frac{\mu _{1} \eta _{\beta 1} - \mu _{2} \eta _{\beta 2}}{\mu _{1} \eta _{\beta 1} + \mu _{2} \eta _{\beta 2}}$

### $T = \frac{2\mu _{1} \eta _{\beta 1}}{\mu _{1} \eta _{\beta 1} + \mu _{2} \eta _{\beta 2}}$
