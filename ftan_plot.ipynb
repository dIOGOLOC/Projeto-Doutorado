```python
fig = plt.figure()
```

```python
data = np.genfromtxt('tmp.3')
x = data[:,0]
y = data[:,1]
z = data[:,2]
```

```python
group = np.genfromtxt('group_2_1.dat')
gper= group[:,0]
gvel= group[:,1]
phase = np.genfromtxt('phase_2_1.dat')
pper= phase[:,0]
pvel= phase[:,1]
```

```python
xi = linspace(min(x), max(x))
yi = linspace(min(y), max(y))
zi = griddata(x,y,z,xi,yi,interp='linear')
```

```python
ax1 = fig.add_subplot
ax1 = plt.plot(gper, gvel,'k',linewidth=1.5, label='Group clean dispersion curve')
ax1 = plt.plot(pper, pvel,'--k',linewidth=1.5, label='Phase clean dispersion curve')
ax1 = plt.contour(xi, yi, zi, 10, linewidths = 0.5, colors = 'k')
ax1 = plt.pcolormesh(xi, yi, zi, cmap = plt.get_cmap('jet'))
ax1 = plt.legend(fancybox=True, shadow=True)
ax1 = plt.colorbar()
plt.savefig('grid_figure_T0.pdf')
```

```python
ny, nx = 600, 600

# Generate a regular grid to interpolate the data.
xi = np.linspace(min(x), max(x), nx)
yi = np.linspace(min(y), max(y), ny)
xi, yi = np.meshgrid(xi, yi)

# Interpolate using delaunay triangularization 
zi = mlab.griddata(x,y,z,xi,yi)
```

```python
ax2 = fig.add_subplot(111)
ax2.plot(gper, gvel,'k',linewidth=1.5, label='Group clean dispersion curve')
#x2.plot(pper, pvel,'--k',linewidth=1.5, label='Phase clean dispersion curve')
#ax2.contour(xi, yi, zi, 10, linewidths = 0.5, colors = 'k')
ax2.pcolormesh(xi,yi,zi)
#plt.scatter(x,y,c=z)
#fig.colorbar(cs,orientation="horizontal")
ax2.set_ylabel('Velocity (sec)')
ax2.yaxis.set_label_position('right')
#ax2.yaxis.tick_right()
plt.show()

#ax2 = plt.savefig('grid_figure_T0.pdf')
#plt.axis([xmin, xmax, ymin, ymax])
```
