#### Comparando as formas de onda provenientes dos métodos iterdecon e pwaveqn pode-se ver que a estabilidade do primeiro método é maior quando comparada ao segundo. Nota-se a presença de alta frequência nas Funções do Receptor calculadas a partir do método pwaveqn, isto faz com que as reflexões da onda P sejam mascaradas por esse ruído de alta frequência. Importante notar que as Funções do Receptor calculadas com o filtro gaussiano (a) igual a 1.0, em ambos os métodos, gera uma forma de onda com um grande conteúdo de alta frequência. Logo o cálculo da Função do Receptor é mais confiável e visualmente possui uma melhor estabilidade quando calculado pelo método iterdecon com um filtro gaussiano (a) igual a 2.5.

<h4 align="center">Calculando o valor do parâmetro do raio (p) para cada evento</h4>

```python
PR_s_km = []
model = TauPyModel(model="iasp91")

for i,j in enumerate(RFPWAVEQNr):
    arrivals = model.get_travel_times(source_depth_in_km=j.stats.sac.evdp,distance_in_degree=j.stats.sac.gcarc, phase_list=["P"])
    arr = arrivals[0]
    PR = arr.ray_param/6371
    PR_s_km.append(PR)
```

```python
for i,j in enumerate(RFPWAVEQNr):
    j.stats.sac['user8']=PR_s_km[i]
    j.write(RFlistPWAVEQNr[i],format='SAC')
```
