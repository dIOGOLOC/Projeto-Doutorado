<h4 aling='justify'>Definindo funções para  baixar os dados, remover a resposta instrumental e visualizar	  o	  sismograma	  em	  várias	  faixas	  de	  frequência	  para	  ver	  a	  relação	  sinal‐ruído	  nessas frequências</h4>

```python
def get_data(network,station,location,channel,starttime,endtime):
    
    station_name = client.get_waveforms(network,station,location,channel,starttime,endtime,attach_response=True)
    
    return station_name
```

```python
def resp_instrumental(station_name,network,station,location,channel,starttime,endtime):
    
    # Download and save instrument response file into a temporary file
    with NamedTemporaryFile() as tf:
        respf = tf.name
        old_iris_client = OldIris_Client()
        # fetch RESP information from "old" IRIS web service, see obspy.fdsn
        # for accessing the new IRIS FDSN web services
        old_iris_client.resp(network,station,location,channel,starttime,endtime,filename=respf)

        # make a copy to keep our original data
        st_orig = station_name.copy()

        # define a filter band to prevent amplifying noise during the deconvolution
        pre_filt = (0.005, 0.006, 45, 55)

        # this can be the date of your raw data or any date for which the
        # SEED RESP-file is valid
        date = starttime

        seedresp = {'filename': respf,  # RESP filename
                    # when using Trace/Stream.simulate() the "date" parameter can
                    # also be omitted, and the starttime of the trace is then used.
                    'date': date,
                    # Units to return response in ('DIS', 'VEL' or ACC)
                    'units': 'DIS'
                    }

        # Remove instrument response using the information from the given RESP file
        station_name.simulate(paz_remove=None, pre_filt=pre_filt, seedresp=seedresp)
```

```python
def vizualize(self):
    fig,axes = mpl.subplots(nrows=3, ncols=5, sharex=True, sharey=False,figsize=(20, 10))
        
    for i,j in enumerate(self):
        axes[i,0].plot(j.times(), j,'k')
        #axes[i,0].set_yticklabels([])
        axes[2,0].set_xlabel('time (s)')
    axes[0,0].set_title('RAW DATA')
    axes[0,0].set_ylabel('BHN')
    axes[1,0].set_ylabel('BHE')
    axes[2,0].set_ylabel('BHZ')

    
    for i,j in enumerate(self):
        j.detrend() #tendêndia linear
        j.taper(type='hamming',max_percentage=0.2) #taper  
        axes[i,1].plot(j.times(), j,'k')
        axes[i,1].set_yticklabels([])
        axes[2,1].set_xlabel('time (s)')
    axes[0,1].set_title('DATA detrend - taper')
    
    for k,l in enumerate(self):
        l.filter('bandpass',freqmin=0.005 ,freqmax=0.01,corners=4,zerophase=True) #filtro passa-banda
        axes[k,2].plot(l.times(), l.data,'k')
        axes[k,2].set_yticklabels([])
        axes[2,2].set_xlabel('time (s)')        
    axes[0,2].set_title('200 to 100 seconds')

    for h,u in enumerate(self):
        u.detrend() #tendêndia linear
        u.taper(type='hamming',max_percentage=0.2) #taper
        u.filter('bandpass',freqmin=0.01 ,freqmax=0.02,corners=4,zerophase=True) #filtro passa-banda
        axes[h,3].plot(u.times(), u.data,'k')
        axes[h,3].set_yticklabels([])
        axes[2,3].set_xlabel('time (s)')        
    axes[0, 3].set_title('100 to 50 seconds')
    
    for v,b in enumerate(self):
        b.detrend() #tendêndia linear
        b.taper(type='hamming',max_percentage=0.2) #taper
        b.filter('bandpass',freqmin=0.02 ,freqmax=0.1,corners=4,zerophase=True) #filtro passa-banda
        axes[v,4].plot(b.times(), b.data,'k')
        axes[v,4].set_yticklabels([])
        axes[2,4].set_xlabel('time (s)')
    axes[0, 4].set_title('50 to 10 seconds')
```
