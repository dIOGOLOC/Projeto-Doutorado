<h3> b) Descreva	 as	  diferenças	  entre	  as	  duas	  seções	  sísmicas.	 </h3> 

<h4 align="justify">A primeira difença marcante entre as seçoes sismicas das funcoes do receptor e a amplitude em relacao as diferentes vagarosidades, pois quanto maior a vagarosidade maior sera a amplitude das Funçoes do Receptor. Observa-se que quando a vagarosidade e igual a 0.08 s/km a amplitude da onda P direta e mais que o dobro da amplitude da onda P direta com a vagarosidade de 0.04 s/km.</h4>


<h4 align="justify">Outra caracteristica interessante na seçao sismica e que com o aumento da vagarosidade a conversao da onda P em S (Ps) na discontinuidade de Moho aumenta em relaçao a chegada da onda P direta. Observa-se devido a proximidade do segundo pico (Ps) em relaçao a linha pontilhada azul (5 segundos). </h4>

<h4 align="justify">Quanto maior a vagarosidade, mais facil pode-se indentificar outras conversoes da Onda P. Por exemplo, para vagarosidades acima de 0.06 s/km pode-se ver claramente a reverberaçoes PpPp. Notavel tambem que para as Funçoes do Receptor para vagarosidades menores que 0.06 s/km a amplitude da conversao de Ps e menor que a conversao em PpPs. Ja para valores maiores que 0.06 s/km, as reflexoes Ps possuem amplitude maior que as reverberaçoes PpPs. </h4>

<h4 align="justify">A largura das Funçoes do Receptor e bem diferenciada entre os dois perfis. Nota-se que quanto maior o filtro gaussiano (a) menor a abertura dos sinais. O Filtro Gaussiano utilizado no calculo das Funçoes do Receptor é um filtro passa-baixa suave, quanto menor a abertura da curva, maior é o conteúdo de frequências eliminadas, porem a utilizaçao deste filtro varia de acordo com o time de estrutura em subsuperficie.</h4>

Nota-se um problema no inicio do sinal, alem de subestimar o tempo de chegada da conversao Ps (referencia a linha pontilhada azul) para um mesmo modelo de velocidade, como visto no primeiro grafico. Ve-se tambem pequenos picos negativos nos primeiros dois graficos apos os sinais positivos. Creio que este problema de instabilidade esteja ligado a quantidade de energia contida no registro do sinal.</h4>
